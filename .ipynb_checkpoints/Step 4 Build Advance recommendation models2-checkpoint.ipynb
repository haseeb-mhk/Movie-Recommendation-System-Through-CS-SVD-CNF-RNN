```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, precision_score, recall_score, f1_score, accuracy_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, Dense, Flatten, Concatenate, Input
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Model

# Load the datasets
ratings = pd.read_csv('Dataset/ratings.csv', encoding='latin-1')
movies = pd.read_csv('Dataset/movies.csv', encoding='latin-1')
users = pd.read_csv('Dataset/users.csv', encoding='latin-1')

# Merge datasets
data = pd.merge(ratings, movies, on='movie_id')
data = pd.merge(data, users, on='user_id')

# Data Preprocessing
# Mapping user_id and movie_id to a sequential index
user_mapping = {user_id: idx for idx, user_id in enumerate(data['user_id'].unique())}
movie_mapping = {movie_id: idx for idx, movie_id in enumerate(data['movie_id'].unique())}

data['user_id'] = data['user_id'].map(user_mapping)
data['movie_id'] = data['movie_id'].map(movie_mapping)

# Normalize the timestamp
min_timestamp = data['timestamp'].min()
max_timestamp = data['timestamp'].max()
data['timestamp_normalized'] = (data['timestamp'] - min_timestamp) / (max_timestamp - min_timestamp)

# Group and pad sequences for RNN
user_sequences = data.groupby('user_id')['movie_id'].apply(list)
ratings_sequences = data.groupby('user_id')['rating'].apply(list)
timestamp_sequences = data.groupby('user_id')['timestamp_normalized'].apply(list)

# Padding sequences to a fixed length
max_seq_length = 10  # Can adjust based on dataset size
padded_users = pad_sequences(user_sequences, maxlen=max_seq_length, padding='post')
padded_ratings = pad_sequences(ratings_sequences, maxlen=max_seq_length, padding='post')
padded_timestamps = pad_sequences(timestamp_sequences, maxlen=max_seq_length, padding='post')

# Combine features (movie_id and timestamp)
X_combined = np.stack([padded_users, padded_timestamps], axis=-1)

# Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(
    X_combined, padded_ratings, test_size=0.2, random_state=42)

# Model Implementation: Recurrent Neural Network (RNN)
input_layer = Input(shape=(max_seq_length, 2))
movie_embedding = Embedding(input_dim=len(movie_mapping), output_dim=50, input_length=max_seq_length)(input_layer[:, :, 0])
timestamp_input = input_layer[:, :, 1:]

# Combine embeddings and timestamp
combined_input = Concatenate()([movie_embedding, timestamp_input])
rnn_layer = SimpleRNN(50, activation='relu')(combined_input)
output_layer = Dense(max_seq_length, activation='linear')(rnn_layer)

model = Model(inputs=input_layer, outputs=output_layer)

# Compile the model
model.compile(optimizer=Adam(learning_rate=0.001), loss='mse', metrics=['mae'])

# Train the model
history = model.fit(X_train, y_train, epochs=10, batch_size=32, validation_data=(X_test, y_test))

# Evaluation
y_pred = model.predict(X_test)

# Reshape for metric computation
y_test_flat = y_test.flatten()
y_pred_flat = y_pred.flatten()

# Calculate evaluation metrics
mae = mean_absolute_error(y_test_flat, y_pred_flat)
rmse = np.sqrt(mean_squared_error(y_test_flat, y_pred_flat))

print(f"MAE: {mae}")
print(f"RMSE: {rmse}")

# Define precision, recall, f1-score, and accuracy (converting predictions to nearest integer)
y_pred_rounded = np.round(y_pred_flat)
precision = precision_score(y_test_flat, y_pred_rounded, average='micro', zero_division=1)
recall = recall_score(y_test_flat, y_pred_rounded, average='micro', zero_division=1)
f1 = f1_score(y_test_flat, y_pred_rounded, average='micro', zero_division=1)
accuracy = accuracy_score(y_test_flat, y_pred_rounded)

print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1-Score: {f1}")
print(f"Accuracy: {accuracy}")

# Visualization of model performance
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.legend()
plt.title('Model Loss Over Epochs')
plt.show()
```

```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, precision_score, recall_score, f1_score, accuracy_score
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import Embedding, GRU, Dense, Flatten, Concatenate, Input, Dropout, Bidirectional
from tensorflow.keras.optimizers import Adam, RMSprop
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.regularizers import l2

# Load the datasets
ratings = pd.read_csv('Dataset/ratings.csv', encoding='latin-1')
movies = pd.read_csv('Dataset/movies.csv', encoding='latin-1')
users = pd.read_csv('Dataset/users.csv', encoding='latin-1')

# Merge datasets
data = pd.merge(ratings, movies, on='movie_id')
data = pd.merge(data, users, on='user_id')

# Data Preprocessing
# Mapping user_id and movie_id to a sequential index
user_mapping = {user_id: idx for idx, user_id in enumerate(data['user_id'].unique())}
movie_mapping = {movie_id: idx for idx, movie_id in enumerate(data['movie_id'].unique())}

data['user_id'] = data['user_id'].map(user_mapping)
data['movie_id'] = data['movie_id'].map(movie_mapping)

# Normalize the timestamp
min_timestamp = data['timestamp'].min()
max_timestamp = data['timestamp'].max()
data['timestamp_normalized'] = (data['timestamp'] - min_timestamp) / (max_timestamp - min_timestamp)

# Cyclic transformation for timestamps (e.g., hour of the day)
data['hour'] = pd.to_datetime(data['timestamp'], unit='s').dt.hour
data['hour_sin'] = np.sin(2 * np.pi * data['hour'] / 24)
data['hour_cos'] = np.cos(2 * np.pi * data['hour'] / 24)

# Group and pad sequences for RNN
user_sequences = data.groupby('user_id')['movie_id'].apply(list)
ratings_sequences = data.groupby('user_id')['rating'].apply(list)
hour_sin_sequences = data.groupby('user_id')['hour_sin'].apply(list)
hour_cos_sequences = data.groupby('user_id')['hour_cos'].apply(list)

# Padding sequences to a fixed length
max_seq_length = 10  # Can adjust based on dataset size
padded_users = pad_sequences(user_sequences, maxlen=max_seq_length, padding='post')
padded_ratings = pad_sequences(ratings_sequences, maxlen=max_seq_length, padding='post')
padded_hour_sin = pad_sequences(hour_sin_sequences, maxlen=max_seq_length, padding='post')
padded_hour_cos = pad_sequences(hour_cos_sequences, maxlen=max_seq_length, padding='post')

# Combine features (movie_id and cyclic timestamp features)
X_combined = np.stack([padded_users, padded_hour_sin, padded_hour_cos], axis=-1)

# Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(
    X_combined, padded_ratings, test_size=0.2, random_state=42)

# Model Implementation: Recurrent Neural Network (RNN)
input_layer = Input(shape=(max_seq_length, 3))

# Separate embeddings for movie_id and cyclic timestamp features
movie_embedding = Embedding(input_dim=len(movie_mapping), output_dim=50, input_length=max_seq_length)(input_layer[:, :, 0])
hour_sin_input = input_layer[:, :, 1:2]
hour_cos_input = input_layer[:, :, 2:3]

# Combine embeddings and timestamp features
combined_input = Concatenate()([movie_embedding, hour_sin_input, hour_cos_input])
bi_gru_layer = Bidirectional(GRU(100, activation='relu', kernel_regularizer=l2(0.01), return_sequences=True))(combined_input)
dropout_layer = Dropout(0.3)(bi_gru_layer)
grnn_layer = GRU(50, activation='relu', kernel_regularizer=l2(0.01))(dropout_layer)
output_layer = Dense(max_seq_length, activation='linear')(grnn_layer)

model = Model(inputs=input_layer, outputs=output_layer)

# Compile the model
model.compile(optimizer=RMSprop(learning_rate=0.001), loss='mse', metrics=['mae'])

# Train the model
history = model.fit(X_train, y_train, epochs=20, batch_size=64, validation_data=(X_test, y_test))

# Evaluation
y_pred = model.predict(X_test)

# Reshape for metric computation
y_test_flat = y_test.flatten()
y_pred_flat = y_pred.flatten()

# Calculate evaluation metrics
mae = mean_absolute_error(y_test_flat, y_pred_flat)
rmse = np.sqrt(mean_squared_error(y_test_flat, y_pred_flat))

print(f"MAE: {mae}")
print(f"RMSE: {rmse}")

# Define precision, recall, f1-score, and accuracy (converting predictions to nearest integer)
y_pred_rounded = np.round(y_pred_flat)
precision = precision_score(y_test_flat, y_pred_rounded, average='micro', zero_division=1)
recall = recall_score(y_test_flat, y_pred_rounded, average='micro', zero_division=1)
f1 = f1_score(y_test_flat, y_pred_rounded, average='micro', zero_division=1)
accuracy = accuracy_score(y_test_flat, y_pred_rounded)

print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1-Score: {f1}")
print(f"Accuracy: {accuracy}")

# Visualization of model performance
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.legend()
plt.title('Model Loss Over Epochs')
plt.show()
```

# creating the model

```python
#embed both users and movies in to 50-dimensional vectors
embedding_size= 200   # embedding size is increases from 50 to 100 for model performance improvement

# Create model
# Set input layers
user_id_input = Input(shape=[1], name='user')
movie_id_input = Input(shape=[1], name='movie')

# Create embedding layers for users and movies
user_embedding = Embedding(output_dim=embedding_size, 
                           input_dim=num_users,
                           input_length=1, 
                           embeddings_initializer="he_normal",
                           embeddings_regularizer=keras.regularizers.l2(1e-6),
                           name='user_embedding')(user_id_input)
movie_embedding = Embedding(output_dim=embedding_size, 
                            input_dim=num_movies,
                            input_length=1, 
                            embeddings_initializer="he_normal",
                            embeddings_regularizer=keras.regularizers.l2(1e-6),
                            name='movie_embedding')(movie_id_input)

# Reshape the embedding layers
user_vector = Reshape([embedding_size])(user_embedding)
movie_vector = Reshape([embedding_size])(movie_embedding)

# Concatenate the reshaped embedding layers
concat = Concatenate()([user_vector, movie_vector])
#concat = Dot(1, normalize=False)([user_vector, movie_vector])

# Combine with dense layers
dense1 = Dense(256)(concat)
dense = Dropout(0.2)(dense1)
y = Dense(1, activation="sigmoid")(dense) # The sigmoid activation forces the rating to between 0 and 1


# compile model
model = Model(inputs=[user_id_input, movie_id_input], outputs=y)
model.compile(loss='mse',  optimizer = 'adam')
 

# Fit the model
history = model.fit(x = [x_train[:,0],x_train[:,1]],
          y = y_train,
          batch_size=256, 
          epochs=6,
          validation_data = ([x_val[:,0], x_val[:,1]], y_val),
          )
     
```

# Recurrent Neural Network 

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader

# Data Preparation
class RatingsDataset(Dataset):
    def __init__(self, data, user_mapping, movie_mapping):
        self.data = data
        self.user_mapping = user_mapping
        self.movie_mapping = movie_mapping

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        user_id = self.user_mapping[self.data.iloc[idx]['user_id']]
        movie_id = self.movie_mapping[self.data.iloc[idx]['movie_id']]
        rating = self.data.iloc[idx]['rating']
        return user_id, movie_id, rating

# Load and preprocess the dataset
ratings = pd.read_csv('Dataset/ratings.dat', sep='::', engine='python', encoding='latin-1', 
                      names=['user_id', 'movie_id', 'rating', 'timestamp'])

# Create user and movie mappings
unique_users = ratings['user_id'].unique()
unique_movies = ratings['movie_id'].unique()
user_mapping = {user: idx for idx, user in enumerate(unique_users)}
movie_mapping = {movie: idx for idx, movie in enumerate(unique_movies)}

# Map users and movies to indices
ratings['user_id'] = ratings['user_id'].map(user_mapping)
ratings['movie_id'] = ratings['movie_id'].map(movie_mapping)

# Split the data
train_data, test_data = train_test_split(ratings, test_size=0.2, random_state=42)

train_dataset = RatingsDataset(train_data, user_mapping, movie_mapping)
test_dataset = RatingsDataset(test_data, user_mapping, movie_mapping)

train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=64, shuffle=False)

# Define the RNN Model
class RNNRecommendationModel(nn.Module):
    def __init__(self, num_users, num_movies, embedding_dim, hidden_dim):
        super(RNNRecommendationModel, self).__init__()
        self.user_embedding = nn.Embedding(num_users, embedding_dim)
        self.movie_embedding = nn.Embedding(num_movies, embedding_dim)
        self.rnn = nn.GRU(embedding_dim * 2, hidden_dim, batch_first=True)
        self.fc = nn.Linear(hidden_dim, 1)

    def forward(self, user_ids, movie_ids):
        user_embeds = self.user_embedding(user_ids)
        movie_embeds = self.movie_embedding(movie_ids)
        x = torch.cat([user_embeds, movie_embeds], dim=1).unsqueeze(1)
        rnn_out, _ = self.rnn(x)
        out = self.fc(rnn_out[:, -1, :])
        return out.squeeze()

# Model Initialization
num_users = len(user_mapping)
num_movies = len(movie_mapping)
embedding_dim = 50
hidden_dim = 128

model = RNNRecommendationModel(num_users, num_movies, embedding_dim, hidden_dim)

# Training the Model
criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model.to(device)

# Training Loop
epochs = 5
for epoch in range(epochs):
    model.train()
    train_loss = 0
    for user_ids, movie_ids, ratings in train_loader:
        user_ids, movie_ids, ratings = user_ids.to(device), movie_ids.to(device), ratings.float().to(device)
        optimizer.zero_grad()
        predictions = model(user_ids, movie_ids)
        loss = criterion(predictions, ratings)
        loss.backward()
        optimizer.step()
        train_loss += loss.item()

    print(f"Epoch {epoch + 1}/{epochs}, Training Loss: {train_loss / len(train_loader):.4f}")

# Evaluation
model.eval()
targets, predictions = [], []
with torch.no_grad():
    for user_ids, movie_ids, ratings in test_loader:
        user_ids, movie_ids, ratings = user_ids.to(device), movie_ids.to(device), ratings.float().to(device)
        preds = model(user_ids, movie_ids)
        targets.extend(ratings.cpu().numpy())
        predictions.extend(preds.cpu().numpy())

rmse = np.sqrt(mean_squared_error(targets, predictions))
mae = mean_absolute_error(targets, predictions)

print(f"RMSE: {rmse:.4f}, MAE: {mae:.4f}")
```
